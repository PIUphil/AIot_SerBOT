```python
import time
from pop import Pilot
```

```python
bot = Pilot.SerBot()
```

```python
# 고정 축(절대좌표) 이동 - 긴급상황
bot.move(0, 50)  # 0 - 각도, 50 - 속도(0~99) -> 무게때문에 속도가 낮으면 무게때문에 안움직일 수도 있음
time.sleep(1)
bot.move(90, 99)
time.sleep(1)
bot.move(180, 99)
time.sleep(1)
bot.move(270, 50)
time.sleep(1)
bot.stop()
```

```python
# 조향 이동 - 일반적인 상황

bot.setSpeed(50)       # 속도 설정 50

bot.forward()        # 앞으로 전진(0도) 속도 50
bot.steering = -0.5    # -1.0 ~ 1.0  (마이너스 -> 왼쪽, 플러스-> 오른쪽)  1: 최대한(약 30도 각도)
time.sleep(1)

bot.backward(99)         # 180도, 이전의 속도 그대로 유지 (move는 속도 지정해줘야함)
bot.steering = 0.5
time.sleep(1)

bot.stop()
```

```python
# joystick 시뮬레이터

bot.joystick()
```

```python
# 자세제어 - 자이로센서, 가속도센서
# 자이로센서
value = bot.getGyro()    # 회전각 측정
print(value)
```

```python
# 가속도센서 - 순간적인 중력의 쏠림을 측정 (충격을 측정. 속도측정 아님.)
# 고정값이 아니라 직접 실험해서 측정해봐야 함
value = bot.getAccel()
print(value)
```
