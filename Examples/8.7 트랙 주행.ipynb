# 데이터 딥러닝

```python
TF=Pilot.Track_Follow(camera=cam)
TF.load_datasets()
```

```python
TF.train(times=5)
```

```python
value=TF.run()
print(value)
```

```python
TF.show()
```

```python
def is_Left(value):
	print(value<0)

value=TF.run(callback=is_Left)
print(value)
```

```python
Car=Pilot.AutoCar()
Car.setSpeed(50)

def drive(value):
	Car.forward()
	steer=value[‘x’]

	if steer > 1:
	        steer=1
	elif steer < -1:
	        steer=-1

	Car.steering=steer*1.5

while True:
	TF.run(callback=drive)
```
